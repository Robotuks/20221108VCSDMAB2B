```python
import matplotlib.pyplot as plt
```

```python
import numpy as np

x = np.arange(0,100)
y = x * 3
z = x ** 2
sin_x = np.sin(x)
cos_x = np.cos(x)
```

```python
fig = plt.figure()
ax = fig.add_axes( (0,0,1,1) )

ax.plot(x, y)
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')

ax.set_title('Plot title')
```

```python
fig = plt.figure()
ax = fig.add_axes( (0,0,1,1) )
# print(z[50])
z[49] = 2000
z[51] = 2700
ax.plot(x, z)

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_title('Plot title')

ax_2 = fig.add_axes( (0.3, 0.5, 0.2, 0.2) ) # x_start, y_start, width, height
ax_2.plot(x, z)
ax_2.set_xlim(46,54)
ax_2.set_ylim(1800,3000)

ax_2.spines[['right', 'top']].set_visible(False)
```

```python
plt.plot(x,sin_x)
plt.plot(x,cos_x)
plt.xlim((0,40))
```

```python
fig, axes = plt.subplots(3,3)
fig.set_dpi(100)
for row in axes:
    row[2].plot(x,z)
    for col in row:
        col.plot(x,y)

fig.tight_layout()
```

```python
fig, axes = plt.subplots(1,2)
axes[0].plot(x,y, c='b', lw=4, ls='--' )
axes[1].plot(x,z, c='r', lw=4 )
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.countplot(data=tips, x='size')
```

```python
sns.countplot(data=tips, x='size', hue='smoker')
```

```python
sns.jointplot(data=tips, x='size', y='total_bill')
```

```python
sns.jointplot(data=tips, x='size', y='total_bill', kind='kde')
```

```python
sns.jointplot(data=tips, x='size', y='total_bill', kind='reg')
```

```python
sns.jointplot(data=tips, x='size', y='total_bill', kind='resid')
```
