### Concat

```python
df_1 = pd.DataFrame(
    {
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
        "C": ["C0", "C1", "C2", "C3"],
    }
)
df_1
```

```python
df_2 = pd.DataFrame(
    {
        "A": ["A4", "A5", "A6", "A7"],
        "B": ["B4", "B5", "B6", "B7"],
        "C": ["C4", "C5", "C6", "C7"],
    }
)
df_2
```

```python
pd.concat( [df_1, df_2] )
```

```python
pd.concat( [df_1, df_2], ignore_index=True )
```

```python
pd.concat( [df_1, df_2], axis=1, )
```

```python
df_1
```

```python
pd.concat( [df_1, pd.Series(["X0", "X1", "X2"], index=["A", "B", "C"]).to_frame().T ], ignore_index=True )
```

```python
df_example = df_1
df_example
```

```python
%%timeit
df_example = df_1
row = pd.Series(["X0", "X1", "X2"], index=["A", "B", "C"]).to_frame().T
for i in range(1000):
    df_example = pd.concat( [df_example, row ], ignore_index=True )
```

```python
%%timeit
df_example = df_1
df_rows = []
for i in range(1000):
    df_rows.append( ["X0", "X1", "X2"] )

df_rows = pd.DataFrame( df_rows, columns=["A", "B", "C"] )
df_example = pd.concat( [df_example, df_rows ], ignore_index=True )
```

### Join

```python
left = pd.DataFrame(
    {
        "key": ["K0", "K1", "K2", "K5"],
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
    }
)

right = pd.DataFrame(
    {
        "key": ["K0", "K1", "K2", "K4"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    }
)
```

```python
left = left.set_index('key')
left
```

```python
right = right.set_index('key')
```

```python
right
```

```python
left.join( right )
```

```python
left.join( right, how='outer' )
```

```python
left.join( right, how='right' )
```

```python
left.join( right, how='inner' )
```

```python
left = pd.DataFrame(
    {
        "key": ["K0", "K1", "K2", "K5"],
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
        "common": [1,2,3,4]
    }
)

right = pd.DataFrame(
    {
        "key": ["K0", "K1", "K2", "K4"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
        "common": [5,6,7,8]
    }
)
```

```python
left = left.set_index('key')
right = right.set_index('key')
left
```

```python
right
```

```python
left.join(right, lsuffix='_left', rsuffix='_right')
```

### Merge

```python
left = pd.DataFrame(
    {
        "key": ["K0", "K1", "K2", "K5"],
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
    }
)

right = pd.DataFrame(
    {
        "key": ["K0", "K1", "K2", "K4"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    }
)
```

```python
left
```

```python
pd.merge( left, right, how='left', on='key' )
```

### Merge asof

```python
df_1=pd.DataFrame({
    'time': [pd.Timestamp('20-01-2023 06:57:00'),
            pd.Timestamp('20-01-2023 06:57:02'),
            pd.Timestamp('20-01-2023 06:57:04'),
            pd.Timestamp('20-01-2023 06:57:06'),],
     
    'price': [12,14,8,7]})

df_1
```

```python
df_2=pd.DataFrame({
    'time': [pd.Timestamp('20-01-2023 06:57:01'),
            pd.Timestamp('20-01-2023 06:57:03'),
            pd.Timestamp('20-01-2023 06:57:05'),
            pd.Timestamp('20-01-2023 06:57:07'),],
     
    'price': [5,7,9,12]})
df_2
```

```python
pd.merge_asof(df_1, df_2, on='time')
```

```python
pd.merge_asof(df_1, df_2, on='time', direction='forward')
```

```python
pd.merge_asof(df_1, df_2, on='time', direction='nearest')
```
